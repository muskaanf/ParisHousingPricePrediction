```python
from sklearn.model_selection import train_test_split

X = df.drop("price", axis = 1)
y = df["price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, shuffle=True)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Training model on Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
y_preds = rfr.predict(X_test)
r2_score(y_test, y_preds)
```

```python
mean_squared_error(y_test, y_preds)
```

```python
# Convert price column to millions
df['price_millions'] = df['price'] / 1000000

# Print updated DataFrame
df
```

```python
df = df.drop("price", axis=1)
```

```python
# Initiating dictionary for story results
scores = {}
```

```python
from sklearn.model_selection import train_test_split

X = df.drop("price_millions", axis = 1)
y = df["price_millions"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Training model on Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

y_preds = rfr.predict(X_test)
r2 = r2_score(y_test, y_preds)
mse = mean_squared_error(y_test, y_preds)

scores["Random Forest Regressor: "] = {'MSE': mse, 'R2_Score': r2}
```

```python
mean_squared_error(y_test, y_preds)
```

```python
# Training model on Decision tree regressor
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)

y_preds = dtr.predict(X_test)

r2 = r2_score(y_test, y_preds)
mse = mean_squared_error(y_test, y_preds)

print("MSE: ", mse)
print("R2 Score: ", r2)

scores["Decision Tree Regressor: "] = {'MSE': mse, 'R2_Score': r2}
```

```python
# Training model on SVR
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train, y_train)
y_preds = svr.predict(X_test)

r2 = r2_score(y_test, y_preds)
mse = mean_squared_error(y_test, y_preds)

print("MSE: ", mse)
print("R2 Score: ", r2)

scores["SVR: "] = {'MSE': mse, 'R2_Score': r2}
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train, y_train)
y_preds = lasso.predict(X_test)

r2 = r2_score(y_test, y_preds)
mse = mean_squared_error(y_test, y_preds)

print("MSE: ", mse)
print("R2 Score: ", r2)

scores["Lasso: "] = {'MSE': mse, 'R2_Score': r2}
```
